```python
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np

t = np.linspace(0,2*np.pi)
x = np.sin(t)

fig, ax = plt.subplots()
ax.axis([0,2*np.pi,-1,1])
l, = ax.plot([],[])

def animate(i):
    l.set_data(t[:i], x[:i])

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=int(len(t)/2))

from IPython.display import HTML
HTML(ani.to_jshtml())
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np


def circle_pt_gen(radius, resolution, xmin, xmax):
    y_coords = []
    x_coords = []
    x_gen_coords = list(np.linspace(xmin, xmax, num=resolution))
    for i in x_gen_coords:
        diff = radius**2 - i**2
        if diff == 0:
            x_coords.append(i)
            y_coords.append(0)
        elif diff > 0:
            diff_sqrt = np.sqrt(diff)
            x_coords.extend([i, i])
            y_coords.extend([-diff_sqrt, diff_sqrt])
    return x_coords, y_coords

x, y = circle_pt_gen(1, 50, -1, 1)

fig, ax = plt.subplots()
ax.axis([-1, 1, -1, 1])
ax.scatter(np.array(x, dtype=float), np.array(y, dtype=float))

fig.set_size_inches(4, 4)
```
